<li><h5>Quantifying Gun Control (ft Dummy Variables)</h5></li> To create metrics based on gun control policies, we built our own dataset called state_gun_control.csv. Unfortunately for our group, this task was too intricate to be trusted to a web scraping script, and we could not find a reliable site that collected the data we needed in a readable format. With the shifting of policies over time, it is also important that we use data within our primary dataset's timeline. We decided to collect data on policies that were enacted by a cut-off date which we specified as the half-way mark in our date-range (Aug 17th, 2015). We collected publicly available information on existing gun control policies with special attention to when they passed and were enforced in each state via their respective ".gov" websites. We divided the 50 states (51 counting Washington DC) between ourselves and manually entered dummy variables to represent the existence of a specific policy in each state. We restricted ourselves to 6 categories to discern whether or not a state had some version of a specified policy, and ended up creating over 300 unique points for our indicator dummy variables. As a general rule, 1 represents a more restrictive version of a policy is in place, 0 means there is a relaxed or non-existent policy. To elaborate further, the categories and descriptions are as follows:
<h4>waiting_period: (Waiting Period)</h4> Describes whether or not there is a waiting period beyond 24 hours between an authorized person (typically a non-felon over 18) purchasing a firearm and legally being allowed to handle or take custody of the firearm. Could apply to rifles, pistols, or both. 1 means there is a policy in place, 0 means none.
<h4>prch_lic_req: (License to purchase/own)</h4> The state has a policy through which one must obtain a license to be able to purchase and or own firearms and or ammunition. 1 means there is a policy requiring such a license, 0 means there is none.
<h4>cc_rstct: (Restrictions on Concealed Carry)</h4> The state bans citizens from carrying firearms in a concealed manner (under the seat of a car, inside a jacket, etc) in public spaces or citizens require a special license to do so. Typically these laws only apply to public spaces, but there are some cases where it may be illegal to conceal a firearm in a private setting as well.
<h4>mag_rstct: (Restrictions on Magazine Capacity)</h4> A state ban or restriction on magazines (*removeable spring loaded vessels containing cartridges that will feed into a firearm as the bolt cycles). This variable specifically pertains to bans on magazines for semi-automatic firearms excluding shotguns. Many states have magazine restrictions on shotguns or specifically shotguns intended for the hunting of waterfowl. Such laws are intended to protect wildlife from unethical hunting practices or other case by case scenarios. Considering the unlikelihood of shotgun tube length restrictions having any impact on firearm related incidents, we opted to ignore such policies and only include bans that restrict the purchase and or possession of magazines designed for semi-automatic gas-operated firearms regardless of the exact cartridge limit enforced.
<h4>oc_rstrct: (Restrictions on Open Carry)</h4> The state bans citizens from carrying firearms openly or citizens require a special license to do so. This specifically pertains to public spaces and not hunting grounds, as there are many states that allow open carry in specified hunting areas or private property but restrict individuals from carrying in public areas (streets, parks, etc). 1 reflects restrictions on open carry in public, 0 reflects what is referred to in many states as "constitutional carry" or an implicit right to carry firearms in public.
<h4>style_rstrct: (Restrictions on Stylistic/Cosmetic Features)</h4> This category is the most encompassing of our dummy variables. It includes states that have their own lists of make, model, or manufacturer bans or a ban against cosmetic features, functional features, or attachments. For example, New Jersey bans adjustable stocks and muzzle-brakes (or recoil compensators). California bans rifles with pistol grips, folding stocks, foregrips, and underbarrel flare launchers. Some states ban specific firearms by name such as the AR15 or FN Hi-Power with similar justification to the feature/cosmetic bans. Regardless, the nature of these laws converge at their intent to eliminate certain firearms such as large handguns and gas-operated semi-automatic rifles (often erroneously referred to as automatic or assault rifles). Due to the Federal ban of automatic-weapons, any states that only ban features or firearms that are already illegal under Federal law (such as fully-automatic firearms or Chinese imports) are excluded from this identifier. A 1 means a state has a policy matching this description and 0 means there are no policies that meet these criteria.

```python
# Imports
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
from matplotlib import pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# shootings_modified.csv is a reduced size version of shootings-2013-2018.csv from Kaggle
# https://www.kaggle.com/jameslko/gun-violence-data
# Columns that are irrelevant or unusable for this project have been dropped to reduce the file size
# The number of shootings in the original dataset is the same as in the updated dataset -- no rows have been dropped

gv_df = pd.read_csv("shootings_modified.csv")
shootings_df = gv_df[['date', 'state', 'n_killed', 'n_injured', 'incident_characteristics', 'n_guns_involved', 'participant_age', 'congressional_district']]
num_shootings = len(shootings_df)

# How big is the dataset?
print(f'Number of incidents in shootings dataset: {num_shootings}')

earliest_date = shootings_df['date'].min()
latest_date = shootings_df['date'].max()

print(f'The shootings dataset ranges from {earliest_date} to {latest_date}')
# The midpoint of the date range is 2015-08-17.

# Display the first few rows of the data
shootings_df.head(3)
```

```python
# state_gun_control.csv is data we have collected ourselves from a legal reference website
# https://www.findlaw.com/state/criminal-laws/gun-control.html and state specific .gov sites.

# Load the data -- policies 
gun_control_df = pd.read_csv("state_gun_control.csv")
gun_control_df['rstrct_lvl'] = gun_control_df.sum(axis=1,numeric_only=True)
gun_control_df.head(3)
```

```python
# Load census data and show first few rows
census_data = pd.read_csv("state_pop_est_2015.csv")
census_data['state'] = census_data.state.replace(' ', '')
```

```python
# Load census data and show first few rows
census_data = pd.read_csv('state_pop_est_2015.csv')
census_area = pd.read_csv('state_area.csv')
drop_rows = 'Puerto Rico','Northern Mariana Islands','United States Virgin Islands','American Samoa','Guam'
for i in drop_rows:
    census_area = census_area[~census_area.state.str.contains(i)]

census_data = census_data.sort_values(by=['state'], axis=0)
census_data = census_data.reset_index().drop(columns='index')
census_area = census_area.sort_values(by=['state'], axis=0)
census_area = census_area.reset_index().drop(columns='index')
census_area.state = census_data.state
# Census data now contains km2 describing state's area
census_data = census_data.merge(census_area, how = 'inner', on= 'state')
census_data['2015_pop_est'] = pd.to_numeric(census_data['2015_pop_est'])
census_data['km2'] = census_data['km2'].str.replace(',','')
census_data['km2'] = pd.to_numeric(census_data['km2'])
census_data['density'] = (census_data['2015_pop_est'])/census_data['km2']

states = census_data.state.tolist()

def add_census_value(df_in, df_pull, col, states):
    df = df_in.copy()
    df[col] = df.state
    for i in states:
        new_col = int(df_pull.loc[(df_pull.state==i)][col].sum())
        df[col] = df[col].replace({i: new_col}, inplace=False)
        
    df_out = df
    return df_out

# combine gun control variables into the dataframe
census_data = pd.concat([census_data, gun_control_df.drop(columns=['state'])], axis=1)
```

```python
# Determine suicides based on number killed == 1; number injured == 0; incident specifically mentions word suicide
# (using 'uicide' to avoid capitalization)
warnings.filterwarnings('ignore')
shootings_df['suicide'] = np.where((shootings_df['incident_characteristics'].str.contains('uicide',na=False))
                                & (shootings_df['n_killed'] == 1) & (shootings_df['n_injured'] == 0) &
                                ~shootings_df['participant_age'].str.contains('1::',na=False), 1, 0)

violent_shootings = shootings_df[shootings_df['n_killed']+shootings_df['n_injured'] != 0]
deadly_shootings = shootings_df[shootings_df['n_killed'] != 0]
```

```python
# Separate shootings_df into non-suicides and suicides
vlnt_df = shootings_df[shootings_df['suicide'] != 1]

aslt_df = census_data.copy()
aslt_df = add_census_value(aslt_df, vlnt_df, 'n_killed', states)
aslt_df = add_census_value(aslt_df, vlnt_df, 'n_injured', states)
aslt_df['n_incd'] = aslt_df['n_killed']+aslt_df['n_injured']

print(f'Proportion shootings that were deadly: {round(len(deadly_shootings)/len(shootings_df),3)}')
# Calculate proportion of deadly shootings that were suicides
census_data.head(10)
```

```python
aslt_per_100k = aslt_df.copy()
aslt_per_100k['n_killed_p100'] = (aslt_df['n_killed']/aslt_df['2015_pop_est'])*10**5
aslt_per_100k['n_injured_p100'] = (aslt_df['n_killed']/aslt_df['2015_pop_est'])*10**5
aslt_per_100k['n_incdnts_p100'] = (aslt_df['n_incd']/aslt_df['2015_pop_est'])*10**5

aslt_per_100k.head(10)
```

```python
# the independent variables set
predictors = aslt_per_100k[['density', 'waiting_period', 'prch_lic_req', 'cc_rstct', 'mag_rstct', 'oc_rstrct', 'style_rstrct']]
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = predictors.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

The above results show that states with concealed carry laws often do not have the other types of restrictions and vice versa. It also appears that there is only a minor correlation between population density and the overall presence of state gun control.

```python
regr = linear_model.LinearRegression()
predictors = ['density', 'waiting_period', 'prch_lic_req', 'cc_rstct', 'mag_rstct', 'oc_rstrct', 'style_rstrct']
X = aslt_per_100k[predictors]
Y = aslt_per_100k['n_killed_p100']
regr.fit(X, Y)

# print the coefficients of the gun control measures as predictors in the multilinear model:
print("Coefficients of gun control measures as predictors of n_killed per 100k:\n")
for i, col in enumerate(predictors):
    print(col + ": " + str(round(regr.coef_[i],3)))
```

```python
print("Differences per average number killed per 100k per state based on restriction level:\n")

for i in range(6):
    leq = aslt_per_100k[aslt_per_100k['rstrct_lvl'] <= i]
    geq = aslt_per_100k[aslt_per_100k['rstrct_lvl'] > i]

    per_state_leq = leq['n_killed_p100'].sum() / len(leq)
    per_state_geq = geq['n_killed_p100'].sum() / len(geq)
    result = round(per_state_geq - per_state_leq,3)
    print(f'<={i} compared with >={i+1} restrictions: Difference of {result}\n')
print('Positive value difference shows that state(s) with more restrictions had fewer deaths.')
```
